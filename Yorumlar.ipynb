```python
import pandas as pd
from sklearn.utils import shuffle
```

```python
data=pd.read_csv('yorumlar_liste.csv',encoding='iso-8859-9')
```

```python
data.loc[data['puan'] > 0.7, 'puan'] = 1.0
```

```python
data.loc[data['puan'] < 0.5, 'puan'] = 0.0
```

```python
df_1 = data[data.puan > 0.9]
```

```python
df_1
```

```python
df_0 = data[data.puan < 0.1]
```

```python
df_0
```

```python
df=pd.concat([df_0,df_1],ignore_index=True)
```

```python
df
```

```python
df=shuffle(df)
```

```python
df
```

```python
df.to_csv('sonyorum.csv',index=False)
```
